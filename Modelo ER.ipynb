```python
import re
def funcQuitaEspacios(cadena):
    if isinstance(cadena, str):

    # Utiliza la funciÃ³n re.sub para reemplazar mÃºltiples espacios por uno solo
        cadena_sin_espacios_extra = re.sub(r'\s+', ' ', cadena).strip().lower()
        return cadena_sin_espacios_extra
    else:
        pass
```

```python
marcas_1, marcas_2, marcas_3 = df_electric['Brand      '].apply(lambda x: funcQuitaEspacios(x)), df_carro_spec['brand'].apply(lambda x: funcQuitaEspacios(x)), df_consumo['Make          '].apply(lambda x: funcQuitaEspacios(x))
```

```python
list_models = []
for model in marcas_1:
    if model not in list_models:
        list_models.append(model)
    else:
        continue

len(list_models)
# list_models = []
for model in marcas_2:
    if model not in list_models:
        list_models.append(model)
    else:
        continue

len(list_models)
# list_models = []
for model in marcas_3:
    if model not in list_models:
        list_models.append(model)
    else:
        continue

len(list_models)
```

```python
index_None = list_models.index(None)
del list_models[index_None]
```
